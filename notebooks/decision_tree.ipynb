```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read train.csv
train = pd.read_csv('train.csv')
train.head()
```

```python
labels = train['Target'].to_numpy()
train = train.drop('Target', axis=1).to_numpy()
```

```python
# standard normalize dataset using sklearn
# wasn't needed as decision tree doesn't need normalization
# but I did it anyways

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train = scaler.fit_transform(train)
```

```python
# split the dataset into train and val
from sklearn.model_selection import train_test_split

train, val, labels_train, labels_val = train_test_split(train, labels, test_size=0.3)
```

```python
# train the model using decision tree
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=10)
clf.fit(train, labels_train)

# accuracy on train and val dataset
print('Train acc: ', clf.score(train, labels_train))
print('Val acc: ', clf.score(val, labels_val))

# loss on train and val dataset
from sklearn.metrics import log_loss

print('Train loss: ', log_loss(labels_train, clf.predict_proba(train)))
print('Val loss: ', log_loss(labels_val, clf.predict_proba(val)))

# precision, recall, f1-score on val dataset
from sklearn.metrics import classification_report

print(classification_report(labels_val, clf.predict(val)))
```

```python
clf2 = DecisionTreeClassifier(max_depth=15)
clf2.fit(train, labels_train)

# accuracy on train and val dataset
print('Train acc: ', clf2.score(train, labels_train))
print('Val acc: ', clf2.score(val, labels_val))

print('Train loss: ', log_loss(labels_train, clf2.predict_proba(train)))
print('Val loss: ', log_loss(labels_val, clf2.predict_proba(val)))

print(classification_report(labels_val, clf2.predict(val)))
```
